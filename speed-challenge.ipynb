```python
import cv2 as cv
import time
import psutil
from matplotlib import pyplot as plt
import numpy as np

all_data = []
cap = cv.VideoCapture('data/train.mp4')                   
count = 0     
success = True 
while success and count < 10:             
    start = time.time()                                                      
    success, image = cap.read()
    if not success: break
    image = image[...,::-1]
    if count % 2 == 0: all_data.append([image])
    else: all_data[int(count / 2)].append(image)
    end = time.time()
    count += 1  
print(len(all_data))
plt.imshow(all_data[0][0], interpolation='nearest')
plt.show()
```
